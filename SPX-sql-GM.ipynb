### Connect to the database

Let us first load the SQL extension and establish a connection with the database


```python
%load_ext sql
```

```python
import csv, sqlite3

con = sqlite3.connect("my_data1.db")
cur = con.cursor()
```

```python
%sql sqlite:///my_data1.db
```

```python
import pandas as pd
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/labs/module_2/data/Spacex.csv")
df.to_sql("SPACEXTBL", con, if_exists='replace', index=False,method="multi")
```

**Note:This below code is added to remove blank rows from table**


```python
#DROP THE TABLE IF EXISTS

%sql DROP TABLE IF EXISTS SPACEXTABLE;
```

```python
%sql create table SPACEXTABLE as select * from SPACEXTBL where Date is not null
```

## Tasks

Now write and execute SQL queries to solve the assignment tasks.

**Note: If the column names are in mixed case enclose it in double quotes
   For Example "Landing_Outcome"**

### Task 1




##### Display the names of the unique launch sites  in the space mission


```python
display_query = 'select distinct Launch_Site from SPACEXTABLE;'
cur.execute(display_query)
cur.fetchall()
```


### Task 2


#####  Display 5 records where launch sites begin with the string 'CCA' 


```python
display_5_query = 'select * from SPACEXTABLE where Launch_Site like "CCA%" limit 5'
cur.execute(display_5_query)
cur.fetchall()
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
display_total_payload = 'select sum(PAYLOAD_MASS__KG_) from SPACEXTABLE where Customer == "NASA (CRS)"'
cur.execute(display_total_payload)
cur.fetchall()
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
display_avg_payload = 'select avg(PAYLOAD_MASS__KG_) from SPACEXTABLE where Booster_Version=="F9 v1.1"'
cur.execute(display_avg_payload)
cur.fetchall()
```

### Task 5

##### List the date when the first succesful landing outcome in ground pad was achieved.


_Hint:Use min function_ 


```python
display_first_success = 'select min(Date) from SPACEXTABLE where Landing_Outcome=="Success (ground pad)"'
cur.execute(display_first_success)
cur.fetchall()
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
df.Landing_Outcome.value_counts()
```

```python
display_boosters = 'select Booster_Version from SPACEXTABLE where Landing_Outcome == "Success (drone ship)" and PAYLOAD_MASS__KG_>4000 and PAYLOAD_MASS__KG_<6000'
cur.execute(display_boosters)
cur.fetchall()
```

### Task 7




##### List the total number of successful and failure mission outcomes


```python
display_success = 'select count(case when Mission_Outcome like "%success%" then 1 end), count(case when Mission_Outcome like "%failure%" then 1 end) from SPACEXTABLE'
cur.execute(display_success)
cur.fetchall()
```

### Task 8



##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
max_payload = 'select Booster_Version from SPACEXTABLE where PAYLOAD_MASS__KG_==(select max(PAYLOAD_MASS__KG_) from SPACEXTABLE)'
cur.execute(max_payload)
cur.fetchall()
```

### Task 9


##### List the records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015.

**Note: SQLLite does not support monthnames. So you need to use  substr(Date, 6,2) as month to get the months and substr(Date,0,5)='2015' for year.**


```python
launches_2015 = 'select substring(Date,6,2),Landing_Outcome, Booster_Version, Launch_Site from SPACEXTABLE where Landing_Outcome == "Failure (drone ship)" and substring(Date,0,5)=="2015"'
cur.execute(launches_2015)
cur.fetchall()
```

### Task 10




##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order.


```python
landing_ordering = 'select Landing_Outcome, count(*) as LANDING_COUNT from SPACEXTABLE where Date>"2010-06-04" and Date<"2017-03-20" group by Landing_Outcome order by LANDING_COUNT desc'
cur.execute(landing_ordering)
cur.fetchall()
```
